a) apresente as informações sobre quais foram as notas dos alunos em cada questão;

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/arquivoCVSparatrabalho.csv")
colunas_calculo = df.columns[2:18]

df.insert(0, 'Aluno', [f'Aluno {i}' for i in range(1, 125)])
```

b) acrescente uma coluna com a nota de cada aluno, que é a soma das notas das questões;

```python
df['Total'] = df[colunas_calculo].sum(axis = 1)
```

c) e apresente o quanto os dados estão dispersos com relação. À nota de cada aluno, facilitando, assim, a análise dos dados (considerar o desvio padrão e a variância).

```python
df['Variancia'] = df[colunas_calculo].var(axis=1)
df['Desvio_Padrao'] = df[colunas_calculo].std(axis=1)

display(df)
```
